```python
def NodeWay(root,tag):
    '''
        Return the dictionary when the fields come from 'node' and 'way'
        arg: root from xml element tree, don't have child
        Return: dictionary
    '''
    
    #Create the dict with all data from the tag 'node' or 'way'
    if tag == 'node':
        #only node has this field
        generic = {'type':tag,'pos':[]}
    else:
        generic = {'type':tag}
    
    for key, value in root.attrib.items():
        if key != 'lat' and key != 'lon':
            generic.update({key:value})
        else:
            #The sort is lat and long in the tag attributes
            generic['pos'].append(float(value))

    return generic
```

```python
def Tag(root):
    '''
        Parse the xml and turn to JSON
        arg1: XML element tree
        return: dictionary
    '''
    
    fields = dict( addr = 'addr', building = 'building', parking = 'parking',
                    contact ='contact', cycleway='cycleway',gym='gym', highway = 'highway',
                    is_in = 'is_in', leisure = 'leisure', museum = 'museum', natural = 'natural',
                    parking = 'parking', payment = 'payment', sidewalk = 'sidewalk')
    
    only_two_levels = ['addr','building','parking','cicleway','is_in','payment']
    
    tag_suffix = root.attrib['k'].split(':')
    
    for key,value in fields.items():
        
        #The tag must stay in valid fields to upload
        if tag_suffix[0].startswith(key) and len(tag_suffix)==2:
            return {value:{tag_suffix[1]:root.attrib['v']}}
        
        elif tag_suffix[0].startswith(key) and len(tag_suffix)==1:
            if key in only_two_levels:
                return None
            
            return {value:root.attrib['v']}
    
```

```python
def XmlToJson(root):
    '''
        Receive the root and change the format to JSON and put in MongoDB
        Arg1: string with in XML format
    '''
    #XML Parse
    root  = ET.fromstring(root)
    
    #for the key 'type' in dict
    tag = root.tag
    
    if tag == 'node' or tag == 'way':
        return NodeWay(root,tag)
    elif tag == 'tag' and len(root.attrib['k'].split(':')) <= 2 :
        #fields with xx:xx:xx more than two levels don't will accepted
        return Tag(root)
```

```python
def PutSameDict(x):
    #The first element must be get to put the others values in the array
    #First all fields needs to be put in the correct key
    aux  = []
    main = x[0]
    for i in x[1:]:
        try:
            aux.append(list(i.items())[0][0])
        except:
            del i

    #Now I know the fields 
    aux = set(aux)

    #Put the values in the same dict
    for key in aux:
        dict_ = {}
        list_ = []
        flag = 0
        for i in x[1:]:
            try:
            #The fields with None return error
                if type(i[key]) == str:
                    list_.append(i[key])
                    flag = 1
                elif type(i[key]) == dict:
                    dict_.update(i[key])
                    flag = 2

            except:
                pass
        #I need tracking the origin of the data
        if flag == 1:
            main[key] = list_
        elif flag == 2:
            main[key] = dict_
        
    return main
```

```python
#the lines must be storage at a variable, after send to the function for put to MongoDB
flag = False
root = None
tag_root = []
count = 0
for i in iter(myfile):
    
    count+=1
    
    decode_ = i.decode('utf-8')
    
    if ('<node' in decode_ or '<way' in decode_) and '/>' in decode_:
        
        db.osm.insert_one(XmlToJson(decode_))
        
    elif ('<node' in decode_ or '<way' in decode_ ) and '/>' not in decode_:
        #Substitute '>' by '/>', so can use xml.elementtree
        root = decode_.replace('>','/>')
        
        #The first part to 'root' are get here, the 'child' comes after
        tag_root.append(XmlToJson(root))
        
        #Flag is a signal that next line is a child
        flag = True
        
    elif ('node>' in decode_ or 'way>' in decode_):
        
        #when drop here, it's need to put all dicts  in same dictionary, and empty the list and throw to MongoDB
        dict_ = PutSameDict(tag_root)
            
        db.osm.insert_one(dict_)
        
        tag_root = [] 
        
        flag=False
        
    elif flag:
        tag_root.append(XmlToJson(decode_))
    
    sys.stdout.write('\rIteracao: {}'.format(count))
```
